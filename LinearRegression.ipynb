```python
train_df = pd.read_csv('house_train.csv')
test_df = pd.read_csv('house_test.csv')
train_df
# test_df
```

```python
train_df.isnull().sum()
```

```python
# FILLING MISSING VALUES IN THE NULL POSITIONS USING MEDIAN AND FILLNA
train_df['Lot_Depth'] = train_df['Lot_Depth'].fillna(train_df['Lot_Depth'].median())
```

```python
# ONE HOT ENCODING ON "LOCATION"
train_df = pd.get_dummies(train_df, columns=['Location'])
test_df = pd.get_dummies(test_df, columns=['Location'])
```

```python
# REQUIREMENTS
X_test = train_df[['Lot_Width','Lot_Depth', 'Year_Sold', 'Year_Built', 'Living_Area', 'Num_Rooms', 'Location_Downtown', 'Location_Rural', 'Location_Suburb']]
y = train_df['Price']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_test,y)
```

```python
predictions = model.predict(X_test)
```

```python
test_ids = test_df['ID']

median_lot_depth_train = train_df['Lot_Depth'].median()
test_df['Lot_Depth'] = test_df['Lot_Depth'].fillna(median_lot_depth_train)

X_test_for_prediction = test_df[['Lot_Width','Lot_Depth', 'Year_Sold', 'Year_Built', 'Living_Area', 'Num_Rooms', 'Location_Downtown', 'Location_Rural', 'Location_Suburb']]

predictions = model.predict(X_test_for_prediction)

submission = pd.DataFrame({
    'ID': test_ids,
    'Price': predictions
})

submission.to_csv('submission.csv', index=False)
```
