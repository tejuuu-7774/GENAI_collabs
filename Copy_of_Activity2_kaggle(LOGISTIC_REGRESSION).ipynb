# OWN CODE

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
train
```

```python
model = LogisticRegression()
model.fit(train[['Input_A', 'Input_B']] , train['Target']);
```

```python
predictions = model.predict(test[['Input_A','Input_B']])
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(train[['Input_A', 'Input_B']])
X_test = scaler.transform(test[['Input_A', 'Input_B']])
model = LogisticRegression(max_iter=1000)
model.fit(X_train, train['Target'])

predictions = model.predict(X_test)
```

```python
submissions = pd.DataFrame({
    'id': test['id'],
    'Target': predictions
})

submissions.to_csv('submission.csv', index=False)
```

# SUBMISSION CODE!

```python
# CODE WITH MORE ACCURACY! DATASET USED
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Features and target
X_train = train[['Input_A', 'Input_B']]
y_train = train['Target']
X_test = test[['Input_A', 'Input_B']]

# Model
model = RandomForestClassifier(
    n_estimators=300,
    max_depth=10,
    random_state=42
)

model.fit(X_train, y_train)
predictions = model.predict(X_test)

# Submission file
submission = pd.DataFrame({
    'id': test['id'],
    'Target': predictions
})

submission.to_csv('submission.csv', index=False)
```
